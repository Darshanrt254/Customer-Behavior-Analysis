```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")

df.sample(5)
```

```python

df.columns
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_" )
df.columns
```

```python
df.isnull().sum()
```

```python
df.review_rating=df.groupby(df.category).review_rating.transform(lambda x : x.median())
```

```python
df.isnull().sum()

df.review_rating
```

```python

df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})

df.columns
```

```python
bins=[0,18,65,150]
labels=["teen","adult","senior"]


df["age_group"]=pd.cut(df.age,bins=bins,labels=labels)
```

```python
df.sample(10)
```

### in frequncy : annually-> 360 

```python
df.frequency_of_purchases
df.frequency_of_purchases.count()
```

```python
frequency_mapping={
    'Fortnightly': 14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-Weekly': 14,
'Annually': 365,
'Every 3 Months': 90
}

df['frequency_of_purchases']=df['frequency_of_purchases'].replace(frequency_mapping)

df.frequency_of_purchases.sample(10)

```
